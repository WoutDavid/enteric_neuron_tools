### COUNTEN Clustering

1 - Enter path to image

2 - Enter path to file containing scripts

3 - Enter pixel density of the image (pixels/micrometer)

4 - Enter sigma to adjust how much smoothing of the image is needed (more background noise requires higher sigma value)

5 - Enter min_samples, the minimum number of neurons in ganglion

```python
neuron_image_path = "/home/david/Documents/enteric_neurons/imgs/1_ME_distal_20X_12WKJune2021_img1_c0_maxIP.tiff"
output_path = "/home/david/Documents/enteric_neurons/output"
script_path = '/home/david/Documents/enteric_neurons/COUNTEN/scripts'
pixel_density = 3.2
sigma = 7
min_samples = 3
```

```python
%load_ext autoreload
%autoreload 2
%reload_ext autoreload

import warnings
warnings.filterwarnings('ignore')

import os
import importlib.util

spec = importlib.util.spec_from_file_location("io",f"{script_path}/io.py")
io = importlib.util.module_from_spec(spec)
spec.loader.exec_module(io)

import javabridge
import bioformats
javabridge.start_vm(class_path=bioformats.JARS)

import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = 10,10
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'


neurons, directory, meta = io.load_TIFF(neuron_image_path, output_path)

javabridge.kill_vm()
```

### Run to display image

```python
plt.imshow(neurons)
plt.axis('off')
```

### Run to process image

```python
spec = importlib.util.spec_from_file_location("processing",f"{script_path}/processing.py")
processing = importlib.util.module_from_spec(spec)
spec.loader.exec_module(processing)

local_maxi, labels, gauss = processing.wide_clusters(neurons, 
                                                sigma=sigma, 
                                                pixel_density=pixel_density,
                                                min_samples=min_samples)

del neurons 
ganglion_prop = processing.segmentation(gauss, local_maxi, labels, meta, directory, save = True)
del gauss
```

### Run to display dataframe 

```python
spec = importlib.util.spec_from_file_location("analysis",f"{script_path}/analysis.py")
analysis = importlib.util.module_from_spec(spec)
spec.loader.exec_module(analysis)

df, dist = analysis.create_dataframe(ganglion_prop, labels, local_maxi, meta, directory, save=True)
df
```
